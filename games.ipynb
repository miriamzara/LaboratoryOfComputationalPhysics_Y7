```python
import numpy as np
from scipy.stats import multivariate_normal

import matplotlib.pyplot as plt

# Define the mean and covariance matrix
mean = [0, 0]
A = np.array([[1, 0.5], [0.5, 1]])
B = np.array([[1, 0], [0, 1]])

# Create grid and multivariate normal
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))
rv = multivariate_normal(mean, A)
plt.contour(x, y, rv.pdf(pos), colors = "black")  # Use contour instead of contourf
rv = multivariate_normal(mean, B)
plt.contour(x, y, rv.pdf(pos), colors = "red")  # Use contour instead of contourf
plt.title('Contour plot of bivariate Gaussian distribution')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(False)  # Optional: turn off grid if present
plt.gca().set_facecolor('white')  # Set white background


plt.show()
```
